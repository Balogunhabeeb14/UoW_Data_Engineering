```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = """
cleaning data
My lecturer is greatly knowledgable and has been very helpful in grasping key terminologies and guiding my through tutorials and
courseworks
That the tutorials were hands on and that we were able to learn new things
The courses very interesting and the lecturer made it seem easy
The exposure and challenges that comes with it
The lecturer, I feel, really magnified my level of understanding on the overall module and previous module like python and SQL, that I
didnt have in my previous modules and I believe it helped me achieve my confidence in skills moving forward.
The teacher is the best out of all my lecturers , I like that .. actively encourages us to ask questions and explains very well , the
interactive lessons are very engaging and helps me to learn better
Using google Colab.
What I like most about Data Engineering is the ability to solve complex problems and build systems that handle large-scale data
efficiently. I enjoy the mix of creativity and technical challenges, especially when working with modern tools and technologies that
constantly push me to learn and grow.
When I have the time to work on the coursework, I enjoy building up to solve the problems. I enjoy working through the codes and
breaking them down bit by bit.

I am very pleased with the lecturer .. did a great job in ensuring we all understand the course
N/A
NA
Really interesting module with a lot of useful information that will help in my career. The module is very well designed and every
tutorial is directly linked with what we see in the lecture, this helps to reinforce our understanding. However, if I must point out
something not that positive in the module is that the lecturers passive tone doesnt contribute to the dynamism of the lectures. So I
suggest if this could be improved this module can be the best module of this semester by far.
The seminars were the most useful, interactive, and enjoyable. Really enhanced my understanding and brought a new found interest
on data engineering
"""

# Tokenize the text
tokens = word_tokenize(text.lower())

# Remove stop words
stop_words = set(stopwords.words('english'))
stop_words.add("module")
filtered_tokens = [word for word in tokens if not word in stop_words]

# Join the filtered tokens back into a string
filtered_text = ' '.join(filtered_tokens)

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(filtered_text)

# Display the generated image
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from textblob import TextBlob
import matplotlib.pyplot as plt

# Sample text (replace with your actual student responses)
text = """
cleaning data
My lecturer is greatly knowledgable and has been very helpful in grasping key terminologies and guiding my through tutorials and
courseworks
That the tutorials were hands on and that we were able to learn new things
The courses very interesting and the lecturer made it seem easy
The exposure and challenges that comes with it
The lecturer, I feel, really magnified my level of understanding on the overall module and previous module like python and SQL, that I
didnt have in my previous modules and I believe it helped me achieve my confidence in skills moving forward.
The teacher is the best out of all my lecturers , I like that .. actively encourages us to ask questions and explains very well , the
interactive lessons are very engaging and helps me to learn better
Using google Colab.
What I like most about Data Engineering is the ability to solve complex problems and build systems that handle large-scale data
efficiently. I enjoy the mix of creativity and technical challenges, especially when working with modern tools and technologies that
constantly push me to learn and grow.
When I have the time to work on the coursework, I enjoy building up to solve the problems. I enjoy working through the codes and
breaking them down bit by bit.
"""

# Split the text into individual responses (assuming each response is separated by a newline)
responses = text.split('\n')

# Perform sentiment analysis and store results
sentiments = []
for response in responses:
    analysis = TextBlob(response)
    polarity = analysis.sentiment.polarity
    sentiments.append(polarity)

# Create a bar chart to visualize sentiment scores
plt.figure(figsize=(10, 6))
plt.bar(range(len(sentiments)), sentiments)
plt.xlabel("Responses")
plt.ylabel("Sentiment Polarity")
plt.title("Sentiment Analysis of Student Responses to the question: What did you most like about Data Engineering?")
plt.ylim(-1, 1)  # Set y-axis limits to -1 and 1 for polarity scores
plt.show()
```
